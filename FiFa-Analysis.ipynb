```python
# a function should be defined to make the process easy when applying 
def remove_star(x):
    x = x.replace('â˜…', '')
    return x
#apply this function to the columns
fifa.IR = fifa.IR.apply(remove_star)
fifa.SM = fifa.SM.apply(remove_star)
fifa['W/F'] = fifa['W/F'].apply(remove_star)
```

```python


def extract_and_replace_numeric(df, columns):
    for column in columns:
        # Extract the first number from each string using a regular expression
        df[column] = df[column].str.extract('(\d+)')
        
        # Convert the column to numeric type
        df[column] = pd.to_numeric(df[column])

# Example usage:
# Assuming 'your_dataframe' is your DataFrame and 'columns_to_process' is a list of column names
columns_to_process = ["LS", "ST","RS","LW","LF","CF","RF","RW","LAM","CAM","RAM","LM","LCM","CM","RCM","RM","LWB","LDM","CDM","RDM","RWB","LB","LCB","CB","RCB","RB","GK"]

# Call the function to extract and replace numeric values in specified columns
extract_and_replace_numeric(fifa, columns_to_process)
fifa
```
